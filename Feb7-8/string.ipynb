## if string is palindrome

```python
def is_palindrome(s):
    s = ''.join(s.split()).lower()  # Remove spaces and convert to lowercase
    i = 0
    j = len(s) - 1
    while i < j:
        if s[i] != s[j]:
            return False
        i += 1
        j -= 1
    return True

# Test cases
print(is_palindrome("racecar"))  # True
print(is_palindrome("hello"))    # False
print(is_palindrome("A man a plan a canal Panama"))  # True (ignoring spaces and case)
```

## Frequency Array

```python
# Frequency array of size 26 for lowercase English letters
s = "hello world"
freq = [0] * 26

for char in s:
    if char.islower():
        index = ord(char) - ord('a')
        freq[index] += 1

print("Frequency array:", freq)
# To map back to letters
for i in range(26):
    if freq[i] > 0:
        print(f"{chr(i + ord('a'))}: {freq[i]}")
```

## Count Frequency of Each Element

```python
# Count frequency of each element in a list
arr = [1, 2, 2, 3, 3, 3, 4]
freq_dict = {}

for num in arr:
    if num in freq_dict:
        freq_dict[num] += 1
    else:
        freq_dict[num] = 1

print("Frequency dictionary:", freq_dict)

# Using collections.Counter for simplicity
from collections import Counter
freq_counter = Counter(arr)
print("Using Counter:", dict(freq_counter))
```

## First Unique Character in a String

```python
# Find the first unique character in a string

# Brute force way: For each character, check if it appears only once
def first_unique_brute(s):
    for i in range(len(s)):
        is_unique = True
        for j in range(len(s)):
            if i != j and s[i] == s[j]:
                is_unique = False
                break
        if is_unique:
            return s[i]
    return None

# Test cases
s1 = "leetcode"
s2 = "loveleetcode"
print("Brute force:", first_unique_brute(s1))  # 'l'

```

```python
# Frequency way: Count frequencies, then find first with count 1
from collections import Counter

def first_unique_freq(s):
    freq = Counter(s)
    for char in s:
        if freq[char] == 1:
            return char
    return None

# Test cases
s1 = "leetcode"
s2 = "loveleetcode"
print("Frequency:", first_unique_freq(s2))     # 'v'
```

```python
# Frequency way: Count frequencies using dict, then find first with count 1
def first_unique_freq(s):
    freq = {}
    for char in s:
        freq[char] = freq.get(char, 0) + 1
    for char in s:
        if freq[char] == 1:
            return char
    return None

# Test cases
s1 = "leetcode"
s2 = "loveleetcode"
print("Frequency:", first_unique_freq(s1))  # 'l'
print("Frequency:", first_unique_freq(s2))  # 'v'
```

```python
# Optimized way: Use a dictionary to track first index and count
def first_unique_optimized(s):
    char_count = {}
    char_index = {}
    for i, char in enumerate(s):
        if char not in char_count:
            char_count[char] = 0
            char_index[char] = i
        char_count[char] += 1
    # Find the minimum index where count == 1
    min_index = float('inf')
    result = None
    for char, count in char_count.items():
        if count == 1 and char_index[char] < min_index:
            min_index = char_index[char]
            result = char
    return result
# Test cases
s1 = "leetcode"
s2 = "loveleetcode"
print("Optimized:", first_unique_optimized(s1)) # None
```

## First Repeating Character in a String

```python
# Find the first repeating character in a string
def first_repeating_char(s):
    seen = set()
    for char in s:
        if char in seen:
            return char
        seen.add(char)
    return None  # No repeating character

# Test cases
s1 = "leetcode"
print("First repeating in 'leetcode':", first_repeating_char(s1))  # 'e'

s2 = "loveleetcode"
print("First repeating in 'loveleetcode':", first_repeating_char(s2))  # 'e'

s3 = "abc"
print("First repeating in 'abc':", first_repeating_char(s3))  # None
```

## Reverse Words in a String

```python
# Reverse the order of words in a string, without leading/trailing spaces
def reverse_words(s):
    s = s.strip()  # Remove leading and trailing spaces
    words = s.split()  # Split and remove extra spaces
    return ' '.join(words[::-1])

# Test cases
s1 = "sky is blue"
print("Reversed:", reverse_words(s1))  # "blue is sky"

s2 = "  hello   world  "
print("Reversed:", reverse_words(s2))  # "world hello"

s3 = "a b c"
print("Reversed:", reverse_words(s3))  # "c b a"
```

## Longest substring witout repeating char

```python
def length_of_longest_substring(s: str) -> int:
    char_set = set()
    left = 0
    max_length = 0

    for right in range(len(s)):
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        char_set.add(s[right])
        max_length = max(max_length, right - left + 1)

    return max_length

# Example usage
print(length_of_longest_substring("abcabcbb"))  # Output: 3
print(length_of_longest_substring("bbbbb"))     # Output: 1
print(length_of_longest_substring("pwwkew"))    # Output: 3
```

```python
def brute_force_longest(s):
    n = len(s)
    max_len = 0
    for i in range(n):
        seen = set()
        for j in range(i, n):
            if s[j] in seen:
                break
            seen.add(s[j])
            max_len = max(max_len, j - i + 1)
    return max_len
s = "abcabcbb"
brute_force_longest(s)
```

```python
def length_of_longest_substring_bruteforce(s: str) -> int:
    def all_unique(sub: str) -> bool:
        return len(set(sub)) == len(sub)

    max_len = 0
    n = len(s)
    for i in range(n):
        for j in range(i+1, n+1):
            if all_unique(s[i:j]):
                max_len = max(max_len, j - i)
    return max_len
s = "abcabcbb"
length_of_longest_substring_bruteforce(s)
```

```python
def length_of_longest_substring_freq(s: str) -> int:
    freq = {}
    left = 0
    max_len = 0

    for right in range(len(s)):
        # Add current character to frequency map
        freq[s[right]] = freq.get(s[right], 0) + 1

        # If duplicate exists, shrink window
        while freq[s[right]] > 1:
            freq[s[left]] -= 1
            left += 1

        # Update max length
        max_len = max(max_len, right - left + 1)

    return max_len

# Example usage
print(length_of_longest_substring_freq("abcabcbb"))  # Output: 3
print(length_of_longest_substring_freq("bbbbb"))     # Output: 1
print(length_of_longest_substring_freq("pwwkew"))    # Output: 3
```

## Longest Substring with character replacement

```python
s="AABABBA" 
k=1

def characterReplacement(s: str, k: int) -> int:
    freq = {}
    left = 0
    max_freq = 0
    max_len = 0

    for right in range(len(s)):
        # Add current char
        freq[s[right]] = freq.get(s[right], 0) + 1
        max_freq = max(max_freq, freq[s[right]])

        # If window invalid, shrink from left
        while (right - left + 1) - max_freq > k:
            freq[s[left]] -= 1
            left += 1

        # Update max length
        max_len = max(max_len, right - left + 1)

    return max_len

# Example usage
print(characterReplacement("AABABBA", 1))  # Output: 4
print(characterReplacement("ABAB", 2))     # Output: 4
```
