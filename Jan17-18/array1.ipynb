## Sum of the array

```python
L = [1,2,3,4,5]
sum = 0
for i in range(0,len(L)):
    sum = sum + L[i]
sum 
```

## Prefix sum = Sum of the running number

```python
L = [1,2,3,4,5]
summ = 0
for i in range(0,len(L)):
    summ = summ + L[i]
    L[i] = summ
L
```

```python
L = [1,2,3,4,5]
for i in range(1,len(L)):
    L[i] = L[i-1] + L[i]
L
```

## Max of the array


```python
L= [5,1,7,2,3]
maxm = -float('inf')

for i in range(0,len(L)):
    if L[i] > maxm:
        maxm = L[i]
maxm
```
